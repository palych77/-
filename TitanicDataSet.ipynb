```python
import matplotlib.pyplot as plt
import pandas as pd
df = pd.read_csv('titanic.csv')
df=df.dropna(subset=['Survived'])
sur = df['Survived']
plt.hist(sur,edgecolor='black', bins =20)
plt.title('Сколько людей выжило на Титанике')
plt.xlabel('Не выжил - 0, выжил - 1')
plt.ylabel('Колличество людей')
plt.show()
```

```python
df = pd.read_csv('titanic.csv')
df=df.dropna(subset=['Sex'])
sex = df['Sex']
plt.hist(sex,edgecolor='black', bins =10)
plt.title('Сколько было мужчин и женщин на титанике')
plt.xlabel('Пол')
plt.ylabel('Колличество людей')
plt.show()
```

```python
df = pd.read_csv('titanic.csv')
df=df.dropna(subset=['Age'])
ages = df['Age']
plt.hist(ages,edgecolor='black', bins =40)
plt.title('Возрастная категория на титанике')
plt.xlabel('Возраст')
plt.ylabel('Колличество людей')
plt.show()
```

```python
df = pd.read_csv('titanic.csv')
df=df.dropna(subset=['SibSp'])
accompanying = df['SibSp']
plt.hist(accompanying,edgecolor='black', bins =20)
plt.title('Сколько было сопровождающих(братьев и сестёр/супруг) на Титанике')
plt.xlabel('Колличество сопровождающих')
plt.ylabel('Колличество людей')
plt.show()
```

```python
df = pd.read_csv('titanic.csv')
df=df.dropna(subset=['Parch'])
family = df['Parch']
plt.hist(family,edgecolor='black', bins =20)
plt.title('Сколько было членов семьи(родителей/детей) на Титанике')
plt.xlabel('Колличество родителей/детей')
plt.ylabel('Колличество людей')
plt.show()
```
